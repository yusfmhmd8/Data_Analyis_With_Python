```python
ax = df.groupby('Coaster_Name')['Speed_Mph'].sum()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind ='barh' ,title = 'Top 10 Coaster Per Total_Speed')
ax.set_xlabel ('#Cosater')
for container in ax.containers:
    ax.bar_label(container)
```

```python
#Top 10 Year Per #Coaster
ax = df.groupby('Year_Introduced')['No'].count()\
.sort_values(ascending=False).head(10)\
.plot (kind = 'bar' , title = 'Top 10 Year Per #Coaster')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Counts')
for container in ax.containers:
    ax.bar_label(container)
#dg = (df.loc[df['Year_Introduced'] == 2018]).groupby('Year_Introduced')['No'].count()
#dg
```

```python
#Top 10 Year Per #Coaster
topYear = df['Year_Introduced'].value_counts().head(10)\
.plot(kind = 'bar',title = "Top 10 Year Per #Coaster" )
topYear.set_xlabel('Year Introduced')
topYear.set_ylabel ('#Cosater')
```

```python
speed = df['Speed_Mph'].plot(kind ='hist',bins = 20,title = 'Coaster Speed (Mph)')
speed.set_xlabel = 'Speed' 
speed.set_ylable = ''
```
